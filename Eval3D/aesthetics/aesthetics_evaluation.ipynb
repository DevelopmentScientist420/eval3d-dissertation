## Aesthetics Evaluation

Scores 3D model aesthetics based on extracted frames from video

```python
import os
from evaluate import get_aesthetic_scores
```

### Initialize variables

```python
# Required input
video_filename = "videos/stone_floor.mp4"

# Optional settings
cache_dir = "sample_results"
n_div = 60
prompt = None  # Use a custom prompt string, or None to infer from filename
```

### Infer get aesthetic scores from evaluate script which returns mean score

```python
if not os.path.exists(video_filename):
    raise FileNotFoundError(f"Video file does not exist: {video_filename}")

score = get_aesthetic_scores(
    video_filename=video_filename,
    cache_dir=cache_dir,
    n_div=n_div,
    prompt=prompt,
)

print(f"Aesthetic score for {video_filename}: {score:.6f}")
```

### Notes
- The cache folder is removed at the end of get_aesthetic_scores
